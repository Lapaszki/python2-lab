Ćwiczenie 3


```python
#Dane
# Produkty w sklepie internetowym
nazwy = ['Laptop', 'Tablet', 'Smartfon', 'Słuchawki', 'Ładowarka',
'Klawiatura', 'Mysz', 'Monitor', 'Kamera', 'Drukarka']
ceny = np.array([3499, 1899, 2499, 349, 89, 249, 129, 1499, 599, 899])
ilosci_magazyn = np.array([15, 42, 78, 156, 312, 89, 201, 23, 45, 34])
oceny_klientow = np.array([4.7, 4.2, 4.5, 4.8, 3.9, 4.1, 4.3, 4.6, 3.8, 3.5])
```

Zadanie 1: Podstawowe statystyki cenowe

```python
print(f"Średnia cena: {ceny.mean():.2f} PLN")
print(f"Najdoroższy produkt: {nazwy[ceny.argmax()]} za {ceny.max()} PLN")
print(f"Najatańszy produkt: {nazwy[ceny.argmin()]} za {ceny.min()} PLN")
```

Zadanie 2: Wartość magazynu


```python
cena_produktu = ceny * ilosci_magazyn 
for nazwa, wartosc in zip(nazwy, cena_produktu):
    print(f"Wartość produktu {nazwa}: {wartosc:.2f} PLN")

print(f"\nWartość magazynu: {np.sum(ceny * ilosci_magazyn):.2f} PLN")
```

Zadanie 3: Analiza cenowa

```python
promocja_swiateczna = ceny * 0.85
ceny_z_vat = ceny * 1.23

print("Promocja_swiateczna:")
print(promocja_swiateczna)

print("Ceny_z_vat:")
print(ceny_z_vat)

print("\nPromocja świąteczna (15% zniżki):")
for nazwa, wartosc in zip(nazwy, promocja_swiateczna):
    print(f"Cena produktu {nazwa} po promocji: {wartosc:.2f} PLN")

print("\nCena z VAT (23%):")
for nazwa, wartosc in zip(nazwy, ceny_z_vat):
    print(f"Cena produktu {nazwa} z VAT: {wartosc:.2f} PLN")

powyzej_1000 = (ceny > 1000).sum()
print(f"\nLiczba produktów powyżej 1000 zł: {powyzej_1000}")
```

Zadanie 4: Filtrowanie

```python
nazwy_array = np.array(nazwy)

maska_ocena = oceny_klientow >= 4.5
print("Produkty z oceną >= 4.5:")
for nazwa, cena in zip(nazwy_array[maska_ocena], ceny[maska_ocena]):
    print(f"Produkt: {nazwa}, Cena: {cena} PLN")
```

```python
maska_cena = ceny < 200
print("\nProdukty poniżej 200 zł:")
for nazwa, ilosc in zip(nazwy_array[maska_cena], ilosci_magazyn[maska_cena]):
    print(f"Produkt: {nazwa}, Ilość w magazynie: {ilosc} szt.")
```

Zadanie 5: Analiza zapasów

```python
nazwy_array = np.array(nazwy)

maska_niski_stan = ilosci_magazyn < 30
print("Produkty z niskim stanem magazynowym (< 30 szt.):")
for nazwa, ilosc in zip(nazwy_array[maska_niski_stan], ilosci_magazyn[maska_niski_stan]):
    print(f"  {nazwa}: {ilosc} szt.")

wartosci = ceny * ilosci_magazyn
wartosc_niski_stan = wartosci[maska_niski_stan].sum()
print(f"\nŁączna wartość tych produktów: {wartosc_niski_stan:.2f} PLN")

wartosc_calego_magazynu = wartosci.sum()
procent = (wartosc_niski_stan / wartosc_calego_magazynu) * 100
print(f"Stanowią {procent:.2f}% wartości całego magazynu")
```

Zadanie 6 (bonus): Analiza korelacji cena–ocena


```python
nazwy_array = np.array(nazwy)
maska_okazja = (oceny_klientow >= 4.5) & (ceny < 1000)
print("Okazje (ocena >= 4.5 i cena < 1000 zł):")
for nazwa, cena in zip(nazwy_array[maska_okazja], ceny[maska_okazja]):
    print(f" {nazwa}: {cena} PLN")

maska_dobra_ocena = oceny_klientow >= 4.0
srednia_dobra = ceny[maska_dobra_ocena].mean()
srednia_slaba = ceny[~maska_dobra_ocena].mean()
print(f"\nŚrednia cena produktów z dobrą oceną >= 4.0: {srednia_dobra:.2f} PLN")
print(f"Średnia cena produktów z słabą oceną < 4.0: {srednia_slaba:.2f} PLN")
```

Ćwiczenie 4: Benchmark — lista vs NumPy

Benchmark

```python
import time
```

```python
rozmiar = 1_000_000
lista = list(range(rozmiar))
tablica = np.array(lista)
```

# Test 1

```python
start=time.perf_counter()
wynik_lista = [x * 2 for x in lista]
czas_lista = time.perf_counter() - start

start = time.perf_counter()
wynik_numpy = tablica * 2
czas_numpy = time.perf_counter() - start

print(f"=== Mnożenie * 2 ({rozmiar:,} elementów) ===")
print(f"Lista: {czas_lista*1000:.1f} ms")
print(f"NumPy: {czas_numpy*1000:.1f} ms")
print(f"NumPy jest {czas_lista/czas_numpy:.0f}× szybszy!")
```

# Test 2

```python
start = time.perf_counter()
suma_lista = sum(lista)
czas_lista = time.perf_counter() - start
start = time.perf_counter()
suma_numpy = tablica.sum()
czas_numpy = time.perf_counter() - start
print(f"\n=== Suma ({rozmiar:,} elementów) ===")
print(f"Lista: {czas_lista*1000:.1f} ms")
print(f"NumPy: {czas_numpy*1000:.1f} ms")
print(f"NumPy jest {czas_lista/czas_numpy:.0f}× szybszy!")
```

# Test 3

```python
start = time.perf_counter()
filtr_lista = [x for x in lista if x > 500000]
czas_lista = time.perf_counter() - start
start = time.perf_counter()
filtr_numpy = tablica[tablica > 500000]
czas_numpy = time.perf_counter() - start
print(f"\n=== Filtrowanie > 500k ({rozmiar:,} elementów) ===")
print(f"Lista: {czas_lista*1000:.1f} ms")
print(f"NumPy: {czas_numpy*1000:.1f} ms")
print(f"NumPy jest {czas_lista/czas_numpy:.0f}× szybszy!")
```

## Wnioski z benchmarku

- NumPy jest **29x** szybszy w mnożeniu
- NumPy jest **5x** szybszy w sumowaniu
- NumPy jest **25** szybszy w filtrowaniu
- Wniosek: przy dużych danych **zawsze** używaj NumPy zamiast list
