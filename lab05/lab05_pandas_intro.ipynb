Zadanie 3a

```python
url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv'
penguins = pd.read_csv(url)

print(penguins.head(10))
print(f"\nShape: {penguins.shape}")
penguins.info()
```

Zadanie 3b

```python
print(penguins.describe())
print("\n--- include all ---")
print(penguins.describe(include='all'))
print(f"\nMediana masy ciała: {penguins['body_mass_g'].median()} g")
```

Zadanie 3c

```python
print(f"Gatunki:\n{penguins['species'].value_counts()}")
print(f"\nWyspy:\n{penguins['island'].value_counts()}")
print(f"\nPłeć:\n{penguins['sex'].value_counts()}")
print(f"\nProcent gatunków:\n{penguins['species'].value_counts(normalize=True).round(3) * 100}")
```

Zadanie 3d

```python
braki = penguins.isna().sum()
print(f"Brakujące wartości:\n{braki}")

braki_sex = penguins['sex'].isna().sum()
procent = braki_sex / len(penguins) * 100
print(f"\nBraki w 'sex': {braki_sex} ({procent:.1f}%)")
```

```python
tips = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')
tips['tip_pct'] = (tips['tip'] / tips['total_bill'] * 100).round(1)

print(f"Średni rachunek: {tips['total_bill'].mean():.2f} zł")
print(f"Średni napiwek %: {tips['tip_pct'].mean():.1f}%")
print(f"\nNajwięcej rachunków wg dnia:\n{tips['day'].value_counts()}")
print(f"\nPalacze:\n{tips['smoker'].value_counts()}")
print(f"\nStatystyki napiwku:\n{tips['tip'].describe()}")
print(f"\nRozkład posiłków:\n{tips['time'].value_counts(normalize=True).round(3) * 100}")
print(f"\nUnikalne rozmiary grup: {tips['size'].nunique()}")
```
