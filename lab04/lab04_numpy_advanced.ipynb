Ćwiczenie 1: Broadcasting — zastosowania biznesowe


Krok 1 — Dane


```python

# Ceny 4 produktów w 3 sklepach
ceny = np.array([[29.99, 31.99, 28.99], # Produkt A
[49.99, 52.99, 47.99], # Produkt B
[14.99, 15.99, 13.99], # Produkt C
[99.99, 104.99, 94.99]]) # Produkt D
sklepy = ['Centrum', 'Galeria', 'Online']
produkty = ['A', 'B', 'C', 'D']
print(f"Kształt macierzy: {ceny.shape}") # (4, 3)

```

Zadanie 1a: Stwórz nowy array z cenami podniesionymi o 5% (inflacja).
Wyświetl wynik.



```python
ceny_po_inflacji = np.round(ceny * 1.05, 2)
print("Ceny po inflacji 5%:")
print(ceny_po_inflacji)
```

Ćwiczenie 3: Analiza danych finansowych

Dane

```python

np.random.seed(2026)
# 8 pracowników, dane roczne
pracownicy = ['Anna', 'Jan', 'Ewa', 'Marek', 'Kasia', 'Piotr', 'Zofia', 'Tomek']
pensje = np.array([5500, 7200, 4800, 9100, 6300, 4200, 8500, 5800])
staz = np.array([3, 8, 2, 15, 5, 1, 12, 4])
oceny = np.array([4.2, 3.8, 4.5, 4.0, 4.7, 3.5, 4.3, 3.9])
# Sprzedaż miesięczna 4 produktów przez 6 miesięcy
produkty_nazwy = ['Laptop', 'Tablet', 'Smartfon', 'Akcesorium']
sprzedaz = np.random.randint(20, 200, size=(4, 6))
ceny_produktow = np.array([3500, 1800, 2500, 150])
```

Zadanie 3a Analiza wynagrodzeń

```python
nazwy = np.array(pracownicy)

srednia = pensje.mean()
mediana = np.median(pensje)
odch = np.round(pensje.std(), 2)

print(f"Średnia pensji: {srednia} zł")
print(f"Mediana pensji: {mediana} zł")
print(f"Odchylenie standardowe: {odch} zł")

print(f"\nPowyżej średniej: {nazwy[pensje > srednia]}")
print(f"Poniżej mediany: {nazwy[pensje < mediana]}")

rozstep = pensje.max() - pensje.min()
iqr = np.percentile(pensje, 75) - np.percentile(pensje, 25)
print(f"\nRozstęp: {rozstep} zł")
print(f"IQR: {iqr} zł")
```

Zadanie 3b

```python
premie = np.where(oceny >= 4.5, 2000, np.where(oceny >= 4.0, 1000, 0))
nowe_pensje = pensje + premie

print("Premie:")
for i in range(len(nazwy)):
    print(f"  {nazwy[i]}: {premie[i]} zł → łącznie {nowe_pensje[i]} zł")

najlepsza = nazwy[premie.argmax()]
print(f"\nNajwyższa premia: {najlepsza}")
```

Zadanie 3c

```python
kolejnosc = np.argsort(pensje)[::-1]

print("Ranking pracowników wg pensji:")
for i, idx in enumerate(kolejnosc, 1):
    print(f"  {i}. {nazwy[idx]} — {pensje[idx]} zł, staż: {staz[idx]} lat")
```

Zadanie 3d

```python
kor_staz_pensja = np.corrcoef(staz, pensje)[0, 1]
kor_staz_ocena = np.corrcoef(staz, oceny)[0, 1]

print(f"Korelacja staż–pensja: {np.round(kor_staz_pensja, 3)}")
print(f"Korelacja staż–ocena:  {np.round(kor_staz_ocena, 3)}")

if abs(kor_staz_pensja) > abs(kor_staz_ocena):
    print("Silniejszy związek: staż–pensja")
else:
    print("Silniejszy związek: staż–ocena")
```

Zadanie 3e

```python
ceny_kol = ceny_produktow.reshape(4, 1)
przychod = sprzedaz * ceny_kol

najlepszy_produkt = produkty_nazwy[przychod.sum(axis=1).argmax()]
najlepszy_miesiac = przychod.sum(axis=0).argmax() + 1

print(f"Najlepszy produkt: {najlepszy_produkt}")
print(f"Najlepszy miesiąc: M{najlepszy_miesiac}")

pierwsze_3 = przychod[:, :3].sum()
ostatnie_3 = przychod[:, 3:].sum()
print(f"\nPrzychód M1–M3: {pierwsze_3} zł")
print(f"Przychód M4–M6: {ostatnie_3} zł")
print(f"Trend: {'rosnący 📈' if ostatnie_3 > pierwsze_3 else 'malejący 📉'}")
```

Ćwiczenie 4: Generowanie danych

Krok 1 — Generowanie danych:

```python
np.random.seed(42)

wiek = np.clip(np.random.normal(35, 10, 100).astype(int), 18, None)
wydatki = np.clip(np.random.normal(500, 200, 100), 0, None)
wizyty = np.random.poisson(5, 100)

print("Dane wygenerowane!")
print(f"Liczba klientów: {len(wiek)}")
```

Krok 2 — Analiza:

```python
# 1. Statystyki opisowe
for nazwa, dane in [('Wiek', wiek), ('Wydatki', wydatki), ('Wizyty', wizyty)]:
    print(f"{nazwa} — średnia: {dane.mean():.1f}, mediana: {np.median(dane):.1f}, "
          f"std: {dane.std():.1f}, min: {dane.min():.1f}, max: {dane.max():.1f}")

# 2. Korelacje
kor_wiek_wydatki = np.corrcoef(wiek, wydatki)[0, 1]
kor_wizyty_wydatki = np.corrcoef(wizyty, wydatki)[0, 1]
print(f"\nKorelacja wiek–wydatki:    {kor_wiek_wydatki:.3f}")
print(f"Korelacja wizyty–wydatki:  {kor_wizyty_wydatki:.3f}")

# 3. Segmentacja (np.select zamiast zagnieżdżonego where — jak sugeruje prowadzący)
sr = wydatki.mean()
sd = wydatki.std()

warunki = [wydatki > sr + sd, wydatki < sr - sd]
wartosci = ['WYSOKI', 'NISKI']
segment = np.select(warunki, wartosci, default='ŚREDNI')

# 4. Ile klientów w każdym segmencie
segmenty, liczby = np.unique(segment, return_counts=True)
print("\nSegmentacja klientów:")
for s, l in zip(segmenty, liczby):
    print(f"  {s}: {l} klientów ({l}%)")

# BONUS: np.diff — zmiana wydatków między kolejnymi klientami
roznice = np.diff(wydatki[:10])
print(f"\nZmiany wydatków (pierwsi 10 klientów): {np.round(roznice, 2)}")
```
