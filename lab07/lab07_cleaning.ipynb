Ćwiczenie 3

```python
df_clean = pd.DataFrame(data)

# Krok 1
df_clean['wynagrodzenie'] = df_clean['wynagrodzenie'].replace('brak', np.nan)
df_clean['wynagrodzenie'] = pd.to_numeric(df_clean['wynagrodzenie'], errors='coerce')

# Krok 2
df_clean = df_clean.drop_duplicates().reset_index(drop=True)

# Krok 3
mediana = df_clean['wynagrodzenie'].median()
df_clean['wynagrodzenie'] = df_clean['wynagrodzenie'].fillna(mediana)

# Krok 4
srednia = round(df_clean['ocena_roczna'].mean(), 2)
df_clean['ocena_roczna'] = df_clean['ocena_roczna'].fillna(srednia)

# Krok 5
df_clean['imie'] = df_clean['imie'].str.strip().str.title()

# Krok 6
df_clean['dzial'] = df_clean['dzial'].str.strip().str.title()
df_clean['dzial'] = df_clean['dzial'].str.replace('Hr', 'HR', regex=False)
df_clean['dzial'] = df_clean['dzial'].str.replace('It', 'IT', regex=False)

# Krok 7
df_clean['data_zatrudnienia'] = pd.to_datetime(df_clean['data_zatrudnienia'], errors='coerce')

# Weryfikacja
print("=== WERYFIKACJA ===")
print(f"Shape: {df_clean.shape}")
print(f"\nDtypes:\n{df_clean.dtypes}")
print(f"\nNaN per kolumna:\n{df_clean.isna().sum()}")
print(f"\nUnikalne działy: {sorted(df_clean['dzial'].unique())}")
print(f"\nPierwsze imiona: {df_clean['imie'].tolist()[:8]}")
```

Pytania biznesowe

```python
print("Średnie wynagrodzenie per dział:")
print(df_clean.groupby('dzial')['wynagrodzenie'].mean().round(2))

print("\nŚrednia ocena per dział:")
print(df_clean.groupby('dzial')['ocena_roczna'].mean().round(2))

print("\nTop 3 najlepiej opłacani:")
print(df_clean.nlargest(3, 'wynagrodzenie')[['imie', 'dzial', 'wynagrodzenie']])

print("\nLiczba pracowników per dział:")
print(df_clean['dzial'].value_counts())
```

Rozszerzenie biznesowe

```python
# 1. Najdłuższy staż
najwczesniejsza = df_clean['data_zatrudnienia'].min()
staz_lat = (pd.Timestamp.today() - najwczesniejsza).days // 365
print(f"Najdłuższy staż: {staz_lat} lat (od {najwczesniejsza.date()})")

# 2. Najwyższa ocena w każdym dziale
print("\nNajlepsza ocena per dział:")
for dzial in df_clean['dzial'].unique():
    idx = df_clean[df_clean['dzial'] == dzial]['ocena_roczna'].idxmax()
    pracownik = df_clean.loc[idx]
    print(f"  {dzial}: {pracownik['imie']} ({pracownik['ocena_roczna']})")

# 3. Korelacja staż vs wynagrodzenie
df_clean['dni_stazu'] = (pd.Timestamp.today() - df_clean['data_zatrudnienia']).dt.days
kor = df_clean[['dni_stazu', 'wynagrodzenie']].dropna().corr().iloc[0, 1]
print(f"\nKorelacja staż–wynagrodzenie: {kor:.3f}")
```
